```python
import pandas as pd
from faker import Faker
import glob
import re

class Anonymizer:
    def __init__(self):
        self.fake = Faker('nl_NL')
        self.bank_account_to_fake_iban = {}
        self.name_to_fake_name = {}
        self.rekening_naam_to_category ={}
        self.omschrijven ={}
        self.straat_to_fake_straat ={}
        self.stad_to_fake_stad ={}
        self.postcode_to_fake_postcode={}

    def _anonymize_bank_account(self, bank_account):
        if bank_account not in self.bank_account_to_fake_iban:
            fake_iban = self.fake.iban()
            self.bank_account_to_fake_iban[bank_account] = fake_iban
        else:
            fake_iban = self.bank_account_to_fake_iban[bank_account]
        return fake_iban

    def _anonymize_name(self, name):
        if name not in self.name_to_fake_name:
            fake_name = self.fake.name()
            self.name_to_fake_name[name] = fake_name
        else:
            fake_name = self.name_to_fake_name[name]
        return fake_name

    def _anonymize_rekening_tegen_naam(self, rekening_naam):
        if rekening_naam not in self.rekening_naam_to_category:
            fake_rekening_naam = self.fake.company()
            self.rekening_naam_to_category[rekening_naam] = fake_rekening_naam
        else:
            fake_rekening_naam = self.rekening_naam_to_category[rekening_naam]
        return fake_rekening_naam

    def _anonymize_straat(self, straat):
        if straat not in self.straat_to_fake_straat:
            fake_straat = self.fake.street_name()
            self.straat_to_fake_straat[straat] = fake_straat
        else:
            fake_straat = self.straat_to_fake_straat[straat]
        return fake_straat

    def _anonymize_stad(self, stad):
        if stad not in self.stad_to_fake_stad:
            fake_stad = self.fake.city()
            self.stad_to_fake_stad[stad] = fake_stad
        else:
            fake_stad = self.stad_to_fake_stad[stad]
        return fake_stad

    def _anonymize_postcode(self, postcode):
        postcode = postcode.replace(" ","")
        if postcode not in self.postcode_to_fake_postcode:
            fake_postcode= self.fake.postcode()
            self.postcode_to_fake_postcode[postcode] = fake_postcode
        else:
            fake_postcode = self.postcode_to_fake_postcode[postcode]
        
        return fake_postcode

    def _modify_text(self, text):
        if isinstance(text, str):
            words = text.lower().split("/")
            modified_words = []
            
            i = 0
            while i < len(words):
                word = words[i].strip()
                if word == "name" and i + 1 < len(words):
                    modified_words.append(word.upper())
                    original_name = words[i + 1].upper()
                    
                    if original_name not in self.name_to_fake_name:
                        fake_name = self.fake.name()  # Generate a new fake name
                        modified_words.append(fake_name.upper())
                        self.name_to_fake_name[original_name] = fake_name  # Add to the dictionary
                    else:
                        fake_name = self.name_to_fake_name[original_name]
                        modified_words.append(fake_name.upper())
                    
                    i += 2  # Skip the next word
                elif word == "iban" and i + 1 < len(words):
                    modified_words.append(word.upper())
                    original_iban = words[i + 1].upper()
                    
                    if original_iban not in self.bank_account_to_fake_iban:
                        fake_iban = self.fake.iban()  # Generate a new fake IBAN
                        modified_words.append(fake_iban.upper())
                        self.bank_account_to_fake_iban[original_iban] = fake_iban  # Add to the dictionary
                        # print(self.bank_account_to_fake_iban)
                    else:
                        fake_iban = self.bank_account_to_fake_iban[original_iban]
                        modified_words.append(fake_iban.upper())
                    
                    i += 2  # Skip the next word
                else:
                    modified_words.append(word.upper())
                    i += 1
            
            return "/".join(modified_words)
        else:
            return text

    def _extract_postcode_street_anonymize(self, text):
        parts = text.split()
        
        # Find the index of the first numeric part
        numeric_index = None
        for i, part in enumerate(parts):
            if part.isnumeric():
                numeric_index = i
                break

        if numeric_index is not None:
            postcode = parts[numeric_index]
            
            # Handle postcode longer than four digits
            if len(postcode) > 4:
                next_part = parts[numeric_index + 1]
                if next_part.isalpha() and len(next_part) >= 2:
                    postcode = postcode[-4:] + next_part[:2]
            
            # Add regex to extract post codes
            postcode_pattern = r'\d{4}\s[a-zA-Z]{2}\b'
            postcode_matches = re.findall(postcode_pattern, text)
            postcode = ", ".join(postcode_matches)
            postcode = postcode.replace(" ","")
          
            # Get or generate fake postcode
            fake_postcode= self._anonymize_postcode(postcode)
        
            # fake_postcode = self.postcode_to_fake_postcode.get(postcode, self.fake.postcode())
            # self.postcode_to_fake_postcode[postcode] = fake_postcode
            
            # Get or generate fake street name
            street_name = " ".join(parts[:numeric_index])
            fake_straatnaam = self.straat_to_fake_straat.get(street_name, self.fake.street_name())
            self.straat_to_fake_straat[street_name] = fake_straatnaam
           

            # Find the index of "%" in parts
            city_start_index = None
            for i, part in enumerate(parts):
                if "%" in part:
                    city_start_index = i + 1
                    break

            # Extract city from parts
            street_name = None
            fake_city = None
            if city_start_index is not None:
                city = " ".join(parts[-1:city_start_index])  # Exclude the postcode
                fake_city = self.stad_to_fake_stad.get(city, self.fake.city())
                self.stad_to_fake_stad[city] = fake_city
            else:
                city = None
            # print(self.postcode_to_fake_postcode)
            # Reconstruct the text with anonymized postcode, street, and city
        
            reconstructed_text = f"{fake_straatnaam} {fake_postcode} {fake_city} %1"
            print(reconstructed_text)
            return reconstructed_text
        else:
            return text



            
    
    def anonymize_dataframe(self, data_df,filename=None):
        data_df = data_df.copy()


        if "rekening_nummer" in data_df.columns:
            data_df['Fake_rekening_nummer'] = data_df.apply(lambda row: pd.NA if pd.isnull(row['rekening_nummer']) else self._anonymize_bank_account(row['rekening_nummer']), axis=1)
        if "rekening" in data_df.columns:
            data_df['Fake_rekening'] = data_df.apply(lambda row: pd.NA if pd.isnull(row['rekening']) else self._anonymize_bank_account(row['rekening']), axis=1)
        if "tegenrekening_nummer" in data_df.columns:
            data_df[f'fake_tegenrekening_nummer'] = data_df.apply(lambda row: pd.NA if pd.isnull(row['tegenrekening_nummer']) else self._anonymize_bank_account(row['tegenrekening_nummer']), axis=1)
        if "rekening_naam" in data_df.columns:
            data_df['Fake_rekening_naam'] = data_df.apply(lambda row: pd.NA if pd.isnull(row['rekening_naam']) else self._anonymize_rekening_tegen_naam(row['rekening_naam']), axis=1)
        if "tegenrekening_naam" in data_df.columns:
            data_df['Fake_tegenrekening_naam'] = data_df.apply(lambda row: pd.NA if pd.isnull(row['tegenrekening_naam']) else self._anonymize_rekening_tegen_naam(row['tegenrekening_naam']), axis=1)
        if "company" in data_df.columns:
            data_df['Fake_company'] = data_df.apply(lambda row: pd.NA if pd.isnull(row['company']) else self._anonymize_rekening_tegen_naam(row['company']), axis=1)
        if "naam" in data_df.columns:
            data_df['Fake_naam'] = data_df.apply(lambda row: pd.NA if pd.isnull(row['naam']) else self._anonymize_rekening_tegen_naam(row['naam']), axis=1)        
        if "straat" in data_df.columns:
            data_df['Fake_straat'] = data_df.apply(lambda row: pd.NA if pd.isnull(row['straat']) else self._anonymize_rekening_tegen_naam(row['straat']), axis=1)        
        if "stad" in data_df.columns:
            data_df['Fake_stad'] = data_df.apply(lambda row: pd.NA if pd.isnull(row['stad']) else self._anonymize_stad(row['stad']), axis=1)  
 
        if filename == "deb.csv" and "straat" in data_df.columns and "postcode" in data_df.columns:
            data_df['Fake_straat'] = data_df.apply(lambda row: pd.NA if pd.isnull(row['straat']) else self._extract_postcode_street_anonymize(row['straat']), axis=1)   
            data_df['Fake_postcode'] = data_df.apply(lambda row: pd.NA if pd.isnull(row['postcode']) else self._anonymize_postcode(row['postcode'].replace(" ", "")), axis=1)  
        if "omschrijving" in data_df.columns:
            data_df['Fake_omschrijving'] = data_df['omschrijving'].apply(self._modify_text)
        
        return data_df

if __name__ == "__main__":
    anonymizer = Anonymizer()

    for filename in glob.glob(r'*.csv'):
        data_df = pd.read_csv(filename, delimiter=';')
        anonymized_df = anonymizer.anonymize_dataframe(data_df,filename)

        processed_filename = f"processed_{filename}"
        anonymized_df.to_csv(processed_filename, index=False)
        print(f"Processed data saved to: {processed_filename}")
```
