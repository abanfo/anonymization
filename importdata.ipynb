## change individual csv files into a workbook with multiple tabs

```python
import pandas as pd

# List of CSV filenames
csv_filenames = ['data1.csv', 'data2.csv', 'data3.csv']  # Add your CSV filenames here

# Output Excel filename
output_excel_filename = 'output.xlsx'

# Create an Excel writer object
excel_writer = pd.ExcelWriter(output_excel_filename, engine='xlsxwriter')

for csv_filename in csv_filenames:
    # Read the CSV into a DataFrame
    df = pd.read_csv(csv_filename)

    # Write the DataFrame to an Excel sheet
    sheet_name = csv_filename[:-4]  # Use the CSV filename without '.csv' as the sheet name
    df.to_excel(excel_writer, sheet_name=sheet_name, index=False)

# Save the Excel file
# excel_writer.save()
# excel_writer.close()  # Close the Excel writer object


      
```
